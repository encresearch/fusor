# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#for 3d Graphs
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
```

## 2. Import CSV file

```python
data=pd.read_csv('TiwanEarthquakeData.csv')
```

# 3. Data Visualization

```python
# Top 5 rows of the Data
print(data[:5])
```

```python
# acquiring height, latitude and tec data for data visualization

height=data["HEIGHT_KM"]
latitude=data["LATITUDE"]
tec=data["TEC"]
```

```python
#x,y=np.meshgrid(height,latitude)
```

```python
x=height
y=latitude
z=tec
```

## 3.1 Plotting in 3D graph for better visualization

```python


fig=plt.figure(figsize=[16,12])

ax=fig.gca(projection='3d')
ax.set_title('TEC during the time of Earquake with respect to Location and Height (2006/05/28, 10:16 UT, Tiwan) ', fontsize=18)

ax.set_xlabel('Height_KM',fontsize=20)
ax.set_ylabel('LONGITUDE',fontsize=20)
ax.set_zlabel('TEC, 10^16 m-2',fontsize=20)
ax.get_yticklabels()[3].set_color("red")
ax.plot_trisurf(x,y,z,linewidth=0, antialiased=False ,cmap=cm.coolwarm)

plt.show()

```
